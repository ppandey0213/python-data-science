```python
import  numpy as np
import pandas as pd
from  sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv("C:\\Users\\pande\\Downloads\\googleplaystore.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
impute = SimpleImputer(missing_values = np.nan , strategy = 'mean')

impute.fit(df.iloc[ : , 2:3 ].values)
```

```python
df.iloc[:,2:3] = impute.transform(df.iloc[:,2:3].values)
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```
