```python
import pandas as pd
import seaborn as sns
df = pd.read_csv("AB_NYC_2019.csv")
df.head()
```

### Histogram

```python
sns.histplot(data = df, x="price")
```

```python
sns.histplot(data = df[df["price"]<500], x="price")
```

```python
sns.histplot(data = df, x="availability_365")
```

### Distribution

```python
sns.displot(df["price"])
```

```python
sns.displot(df["availability_365"])
```

```python
sns.displot(df["availability_365"], kde=False, stat="probability")
```

```python
sns.displot(df["reviews_per_month"])
```

### Boxplot

```python
sns.boxplot(df["price"])
```

```python
sns.boxplot(df["availability_365"])
```
