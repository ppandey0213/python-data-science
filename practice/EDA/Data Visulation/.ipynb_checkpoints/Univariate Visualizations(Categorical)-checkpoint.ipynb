```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("AB_NYC_2019.csv")
df.head()
```

### Categorical

#### Bar Graph

```python
sns.countplot(data=df, x="neighbourhood_group", palette="Set2")
```

```python
sns.countplot(data=df, x="room_type", palette="Set2")
```

```python
sns.countplot(data=df, x="neighbourhood_group",hue="room_type" ,palette="Set2")
```

#### Pie Chart

```python
df["room_type"].value_counts().plot(kind = "pie", autopct = "%.2f")
```

```python
df["neighbourhood_group"].value_counts().plot(kind = "pie", autopct = "%.2f")
```

```python
df["neighbourhood_group"].value_counts().plot(kind = "bar")
```
